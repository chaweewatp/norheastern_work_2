```python
s='zyxwvutsrqponmlkjihgfedcba'
```

```python
alphabetical='abcdefghijklmnopqrstuvwxyz'
longest_length=0
longest_substring='z'
for ind1 in range(0,len(s)):
    substring=s[ind1]
    length_substring=len(substring)
    remaining_substring=s[ind1+1:]
    
    for str2 in remaining_substring:
        if alphabetical.find(str2) >= alphabetical.find(substring[-1]):
            substring=substring+str2
            length_substring +=1
            if length_substring > longest_length:
                longest_length=length_substring
                longest_substring=substring
        else:
            break
print('Longest substring in alphabetical order is: ' + longest_substring)
```
